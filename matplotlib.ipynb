Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

```python
from matplotlib import pylab
from matplotlib.pyplot import subplot

print(pylab.__version__)
```

```python
import numpy as np

x = np.linspace(0, 10, 25)
y = x * x + 2
print(x)
print(y)
print(np.array([x, y]).reshape(25, 2))
```

```python
pylab.plot(x, y, 'ro--')
```

```python
pylab.subplot(1, 2, 1)
pylab.plot(x, y, 'r--')
pylab.title('X vs Y')
pylab.subplot(1,2,2)
pylab.plot(y, x, 'g*-')
pylab.title('Y vs X')
pylab.suptitle('Subplots Example')
pylab.show()
```

```python
import matplotlib.pyplot as plt

plt.subplot(2, 1, 1)  # Top plot
plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Top Plot")

plt.subplot(2, 1, 2)  # Bottom plot
plt.plot([1, 2, 3], [6, 5, 4])
plt.title("Bottom Plot")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.subplot(2, 2, 1)
plt.title("Plot 1")
plt.plot([1, 2, 3], [1, 2, 3])

plt.subplot(2, 2, 2)
plt.title("Plot 2")
plt.plot([1, 2, 3], [1, 4, 9])

plt.subplot(2, 2, 3)
plt.title("Plot 3")
plt.plot([1, 2, 3], [1, 8, 27])

plt.subplot(2, 2, 4)
plt.title("Plot 4")
plt.plot([1, 2, 3], [1, 16, 81])

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Modern Matplotlib Example")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0.5, 0.1, 0.8, 0.8])
ax.plot(x, y, 'r')
```

```python
x= np.linspace(0, 10, 100)
y= np.sin(x)
fig, axes = plt.subplots()
axes.set_xlabel('X-axis')
axes.set_ylabel('Y-axis')
axes.set_title('Simple Plot')
axes.plot(x, y, 'r')
```

```python
x = np.linspace(0, 10, 25)
y = x * x + 2
fig = plt.figure(figsize=(16,9),dpi=300)
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
ax1.plot(x, y, 'ro--')
ax2.plot(x,y,'g*-')
```
