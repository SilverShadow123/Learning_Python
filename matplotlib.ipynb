```python
from matplotlib import pylab
from matplotlib.pyplot import subplot

print(pylab.__version__)
```

```python
import numpy as np

x = np.linspace(0, 10, 25)
y = x * x + 2
print(x)
print(y)
print(np.array([x, y]).reshape(25, 2))
```

```python
pylab.plot(x, y, 'ro--')
```

```python
pylab.subplot(1, 2, 1)
pylab.plot(x, y, 'r--')
pylab.title('X vs Y')
pylab.subplot(1,2,2)
pylab.plot(y, x, 'g*-')
pylab.title('Y vs X')
pylab.suptitle('Subplots Example')
pylab.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 25)

fig, axes = plt.subplots()

axes.set_xlabel('X-axis')
axes.set_ylabel('Y-axis')
axes.set_title('Simple Plot')

axes.plot(x, x**2, label='y = x²')
axes.plot(x, x**3, label='y = x³')

axes.legend(loc='upper left')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
fig, Axes = plt.subplots(1, 2, figsize=(10, 5))

Axes[0].plot(x, x**2, x, x**3, lw=2)
Axes[1].plot(x, x**2, x, x**3, lw=2)

Axes[0].grid(True)
Axes[1].set_ylim([0, 60])
Axes[1].set_xlim([2, 5])
Axes[1].grid(True)

plt.show()
```

Other 2D Graph


```python
import matplotlib.pyplot as plt
import numpy as np
n=np.array([0,1,2,3,4,5])
fig, axes = plt.subplots(1,4, figsize=(16,5))
axes[0].set_title('scatter')
axes[0].scatter(n, n+0.25*np.random.randn(len(n)))
axes[1].set_title('step')
axes[1].step(n,n**2, lw=2)
axes[2].set_title('bar')
axes[2].bar(n,n**2,align='center', color='b')

axes[3].set_title('fill_between')
axes[3].fill_between(n,n**2,n**3,color='r')
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([0.0, 0.0, 0.6, 0.6], polar=True)

theta = np.linspace(0, 2*np.pi, 100)
r = np.abs(np.sin(3*theta))  # radius as a function of angle
ax.plot(theta, r)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create data
x = np.linspace(0, 2*np.pi, 200)   # angle or x-values
y = np.sin(3 * x)                  # function for both systems

# Create figure with 1 row and 2 columns
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# --- 1️⃣ Cartesian Plot ---
axes[0].plot(x, y, color='dodgerblue', lw=2)
axes[0].set_title("Cartesian Plot (x, y)")
axes[0].set_xlabel("X-axis")
axes[0].set_ylabel("Y-axis")
axes[0].grid(True)

# --- 2️⃣ Polar Plot ---
polar_ax = plt.subplot(1, 2, 2, projection='polar')
polar_ax.plot(x, np.abs(y), color='crimson', lw=2)
polar_ax.set_title("Polar Plot (θ, r)")

plt.tight_layout()
plt.show()
```

```python
n= np.random.randn(100000)
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].set_title('Default Histogram')
axes[0].hist(n)
axes[1].set_title('Cumulative detailed histogram')
axes[1].hist(n, cumulative=True, bins=8)
```

```python
import  matplotlib.cm as cm
delta = 0.025
x = np.arange(-3, 3,delta)
y = np.arange(-2, 2, delta)
X, Y = np.meshgrid(x,y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X-1)**2 - (Y-1)**2)
Z = (Z1 - Z2) * 2
fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline=1, fontsize=10)
ax.set_title('Contour Map')
```
